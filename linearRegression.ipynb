```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
data = pd.read_csv('Data.csv')
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
```

```python
x_mean = np.mean(X)
y_mean = np.mean(Y)

nr = dr = 0
for i,j in zip(X,Y):
	nr+=(i-x_mean)*(j-y_mean)
	dr+=(i-x_mean)**2

param1 = nr/dr
param0 = y_mean - param1*x_mean
Y_pred = param1*X + param0
```

```python
plt.scatter(X, Y, marker='o', s=70, c='g')
plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='m', linewidth = 6)
plt.xlabel("Weight", fontsize = 20)
plt.ylabel("Height (In inches)", fontsize = 20)
plt.show()
```

```python
print("Slope - %1.4f\nCoefficient - %1.4f\n" %(param1, param0))

#function for finding output
def Value(x):
    y = param1*x + param0
    return y


x_val = float(input("Enter the Weight(in KGs) - "))
print("Height for this Weight is - %1.3f" %(Value(x_val)))
```
